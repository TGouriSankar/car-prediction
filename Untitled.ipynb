### Model

```python
x = car_price.drop(['Selling_Price'],axis=1)
y = car_price['Selling_Price']
```

```python
x.shape,y.shape
```

```python
x.head()
```

```python
from sklearn.preprocessing import OneHotEncoder
```

```python
ohe = OneHotEncoder(sparse=False,handle_unknown='ignore')
```

```python
x.iloc[:,:3]
```

```python
x_dup = pd.DataFrame(ohe.fit_transform(x.iloc[:,:3]),columns=ohe.get_feature_names_out())
x_dup
```

```python
x.drop(["Fuel_Type","Seller_Type","Transmission"],1,inplace=True)
x
```

```python
x.reset_index(inplace=True)
x_dup.reset_index(inplace=True)
```

```python
x.head()
```

```python
x = pd.concat([x,x_dup],1)
x.drop(['index'],1,inplace=True)
x
```

```python
x
```

```python
### Feature Importance

from sklearn.ensemble import ExtraTreesRegressor
import matplotlib.pyplot as plt
regressor = ExtraTreesRegressor()
regressor.fit(x,y)
```

```python
print(regressor.feature_importances_)
```

```python
#plot graph of feature importances for better visualization
feat_importances = pd.Series(regressor.feature_importances_, index=x.columns)
feat_importances.nlargest(5).plot(kind='barh')
plt.show()
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(x,y,test_size=0.2,random_state=20)
```

```python
x_train.shape,x_test.shape,y_train.shape,y_test.shape
```

```python
model = RandomForestRegressor()
model.fit(x_train,y_train)
```

```python
model.predict(x_test)
```

```python
model.score(x_train,y_train)
```

```python
model.score(x_test,y_test)
```
